```python
# 1. INSTALL COMPATIBILITY LAYER
!pip install tf_keras

# 2. MOUNT GOOGLE DRIVE
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# 3. IMPORT LIBRARIES
import tf_keras as tk
from PIL import Image, ImageOps
import numpy as np

# 4. CONFIGURE DISPLAY SETTINGS
np.set_printoptions(suppress=True)

# 5. LOAD THE MODEL
model = tk.models.load_model("/content/drive/MyDrive/model_files/keras_model.h5", compile=False)

# 6. LOAD THE LABELS
class_names = open("/content/drive/MyDrive/model_files/labels.txt", "r").readlines()

# 7. PREPARE THE DATA ARRAY
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

# 8. LOAD AND PROCESS THE TEST IMAGE
image = Image.open("/content/drive/MyDrive/model_files/test⁠.JPG").convert("RGB")
```

```python
prediction = model.predict(data)

# Determine the class and confidence
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

print("Class:", class_name[2:].strip())
print("Confidence Score:", round(float(confidence_score), 4))
```
